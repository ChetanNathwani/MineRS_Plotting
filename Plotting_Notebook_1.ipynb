# Functions

Creating a function can make life much easier when you want to perform a complex calculation many times

```python
# Define the Ti in zircon thermometer (with pressure correction from Loucks JPet paper)

# Define NAME OF FUNCTION(VARIABLES):
def T_calc(Ti_zrc, P, aTiO2, aSiO2):
    # Do stuff below this line
    T_K = (-4800+(0.4748*(P-1000)))/(np.log10(Ti_zrc)-5.5711-np.log10(aTiO2)+np.log10(aSiO2))
    T_C = T_K - 273.15
    # End of function!
    return(T_C) # Value to return

# Now we can easily calculated Ti in zircon temperatures for all of our data!
# And we are also free to select whichever input parameters we want
data['Temperature'] = T_calc(data['Ti49_ppm'], 200, 0.7, 1.0)
print(data['Temperature'])

# We can also more simply perform calculations on pandas dataframes as follow:
data['Eu/Eu*'] = (data['Eu151_ppm']/0.058)/(np.sqrt((data['Sm147_ppm']/0.153)*(data['Gd157_ppm']/0.2055)))
print(data['Eu/Eu*'])
```

Scatterplots are made using matplotlib's scatter function

https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.scatter.html

```python
import matplotlib.pyplot as plt

# scatter(x_column, y_column)
plt.scatter(data['Temperature'], data['Eu/Eu*'])
plt.show() # DIsplay plot

# WE can adjust lots and lots of parameters, you can find out more in the documentation
plt.scatter(data['Temperature'], data['Eu/Eu*'], c = 'red', marker = 's', edgecolors = 'black')
plt.show()

# Still not a useful plot...
```

Adding further changes to the plot can be done very simply just with extra lines of code (like most coding languages)

```python
plt.scatter(data['Temperature'], data['Eu/Eu*'], c = 'red', marker = 's', edgecolors = 'black')

# Labels
plt.xlabel('Temperature (C)')
plt.ylabel('Eu/Eu*')
plt.title('My new plot!')

# Axes limits
plt.xlim([700,1000])
plt.ylim([0.1,0.7])

plt.show()
```

Let's also now try to make subplots. 

```python
# Fig is the entire figure, axs defines a list of individual axes for each subplot

fig, axs = plt.subplots(1,2, figsize = (8,4))

# Plot on first subplot
axs[0].scatter(data['Temperature'], data['Eu/Eu*'], c = 'red', marker = 's', edgecolors = 'black')
axs[1].scatter(data['Hf177_ppm'], data['Eu/Eu*'], c = 'red', marker = 's', edgecolors = 'black')

# Notice that with subplots the axis commands have a "set_" before them
axs[0].set_xlabel('Temperature (C)')
axs[0].set_ylabel('Eu/Eu*')
axs[1].set_xlabel('Hf (ppm)')

fig.tight_layout() # This optimises space and often works quite well.
plt.show()
```

Plot different units as different colours

```python
groups = data.groupby('Unit2')

for name, group in groups:
    plt.scatter(group['Hf177_ppm'], group['Eu/Eu*'], label = name)
    
plt.legend() # Plot a legend
plt.show()
```
